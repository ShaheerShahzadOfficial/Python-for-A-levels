```python
#  part 1(a)
global array
global newTopTenList


array = [[""] * 2 for i in range(10)]

newTopTenList = [[""] * 2 for i in range(11)]



#  part 1(b)

def ReadHighScores():
    Filename = "HighScore.txt"
    file = open(Filename, "r")
    file_to_be_read = open(Filename, "r")


    for i in range(len(file.readlines())//2):

        playerName = file_to_be_read.readline()
        playerScore = file_to_be_read.readline()
        # print(playerName,i,playerScore)  
        array[i][0] = playerName.split("\n")[0]
        array[i][1] = int(playerScore.split("\n")[0])
    print(array, "array")
    file_to_be_read.close()

# # part 1(c)



def OutputHighScores():

    for i in range(len(array)):

        print(array[i][0], array[i][1])


# # part 1 (d)(i)


ReadHighScores()


OutputHighScores()


# # part 1(e)(i)


name = input("Enter Name of the player, three character only")

while len(name) != 3:

    name = input("Enter Name of the player, three character only")


score = int(input("Enter the score that must be between 1 and 100000"))

while score >= 1 and score <= 100000:

    score = int(input("Enter the score that must be between 1 and 100000"))


# # part 1(e)(ii)



def appendPlayer(name, score):

    Filename = "HighScore.txt"

    file = open(Filename, "r")

    # file_content_length = len(file.readlines())

    # print(file.readlines())

    file_to_be_read = open(Filename, "r")

    for i in range(11):
     if i < 10 :
        PLAYERNAME = file_to_be_read.readline()
        PLAYERScore = file_to_be_read.readline()
        newTopTenList[i][0] = PLAYERNAME.split("\n")[0]

        newTopTenList[i][1] = int(PLAYERScore.split("\n")[0])
     else:
        
        newTopTenList[i][0] = name.upper()
        newTopTenList[i][1] = score
    print(newTopTenList)    



appendPlayer(name, score)



# def appendPlayerInNewList(name, score):

#     Filename = "HighScore.txt"

#     file = open(Filename, "r")

#     # file_content_length = len(file.readlines())

#     # print(file.readlines())

#     file_to_be_read = open(Filename, "r")

#     lastIndex = 0

#     for i in range(len(file.readlines())):

#         content = file_to_be_read.readline()

#         newTopTenList[i][0] = content.split(" ")[0]

#         newTopTenList[i][1] = content.split(" ")[1].split("\n")[0]

#         lastIndex = i

#     newTopTenList[lastIndex + 1][0] = name.upper()

#     newTopTenList[lastIndex + 1][1] = score



# appendPlayerInNewList(name, score)


# OutputHighScores()



def bubbleSort():

    limit = len(newTopTenList) - 1

    swapped = False

    for x in range(limit):

        for y in range(limit - x):

            if newTopTenList[y + 1][1] > newTopTenList[y][1]:

                swapped = True

                newTopTenList[y + 1], newTopTenList[y] = (
                    newTopTenList[y],
                    newTopTenList[y + 1],
                )

        if not swapped:

            return

    print(newTopTenList)
    return



bubbleSort()



def WriteTopTen():

    FileName = "NewHighScore.txt"


    file = open(FileName, "w")

    for i in range(len(newTopTenList)-1):
        file.write(f"{newTopTenList[i][0]}  {newTopTenList[i][1]}")
        if i != len(newTopTenList) -1 : file.write('\n')
    file.close()


WriteTopTen()
```
